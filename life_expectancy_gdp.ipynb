```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("all_data.csv")
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
data_chile = data[data.Country == 'Chile']
data_china = data[data.Country == 'China']
data_germany = data[data.Country == 'Germany']
data_mexico = data[data.Country == 'Mexico']
data_usa = data[data.Country == 'United States of America']
data_zimbabwe = data[data.Country == 'Zimbabwe']
```

```python
plt.plot(data_chile['Year'], data_chile['Life expectancy at birth (years)'])
plt.plot(data_china['Year'], data_china['Life expectancy at birth (years)'])
plt.plot(data_germany['Year'], data_germany['Life expectancy at birth (years)'])
plt.plot(data_mexico['Year'], data_mexico['Life expectancy at birth (years)'])
plt.plot(data_usa['Year'], data_usa['Life expectancy at birth (years)'])
plt.plot(data_zimbabwe['Year'], data_zimbabwe['Life expectancy at birth (years)'])
plt.legend(['Chile', 'China', 'Germany', 'Mexico', 'United States', 'Zimbabwe'])
plt.title('Life Expectancy by Country 2000-2015')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth')
plt.savefig('life_expectancy_2000_2015.png', facecolor='white', transparent=False)
```

```python
plt.plot(data_chile['Year'], data_chile['GDP'])
plt.plot(data_china['Year'], data_china['GDP'])
plt.plot(data_germany['Year'], data_germany['GDP'])
plt.plot(data_mexico['Year'], data_mexico['GDP'])
plt.plot(data_usa['Year'], data_usa['GDP'])
plt.plot(data_zimbabwe['Year'], data_zimbabwe['GDP'])
plt.legend(['Chile', 'China', 'Germany', 'Mexico', 'United States', 'Zimbabwe'], loc = 'upper left', fontsize = 'small')
plt.title('GDP by Country 2000-2015')
plt.xlabel('Year')
plt.ylabel('GDP (USD)')
plt.savefig('GDP_2000_2015.png', facecolor='white', transparent=False)
```

```python
GDP_life_exp_graphs = sns.FacetGrid(data, col = 'Country', col_wrap = 3, hue = 'Country', sharex = False, sharey = False)
GDP_life_exp_graphs.map(sns.scatterplot, 'Life expectancy at birth (years)', 'GDP').add_legend().set_axis_labels('Life expectancy at birth (years)', 'GDP (USD)')
GDP_life_exp_graphs.fig.suptitle('Life Expectancy vs GDP by Country')
GDP_life_exp_graphs.fig.subplots_adjust(top=0.9)
GDP_life_exp_graphs.savefig('GDP_life_exp_graphs.png', facecolor='white', transparent=False)
```
